```python
#1.
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi
def main():
    weight = float(input("Please enter your weight in kilograms: "))
    height = float(input("Please enter your height in meters: "))
    bmi = calculate_bmi(weight, height)
    print(f"Your BMI is: {bmi:.2f}")
if __name__ == "__main__": main()

#2.
def check_eligibility(weight, height, sex):
    if sex.lower() == "male":
        if 62 <= height <= 80 and 150 <= weight <= 280:
            return "You are eligible for the study."
        else:
            return "You are not eligible for the study. Please check the height and weight thresholds for men."
    elif sex.lower() == "female":
        if 59 <= height <= 71 and 100 <= weight <= 180:
            return "You are eligible for the study."
        else:
            return "You are not eligible for the study. Please check the height and weight thresholds for women."
    else:
        return "Invalid input for sex. Please enter 'male' or 'female'."
def main():
    weight = float(input("Please enter your weight in pounds: "))
    height = float(input("Please enter your height in inches: "))
    sex = input("Please enter your sex (male/female): ")
    eligibility_message = check_eligibility(weight, height, sex)
    print(eligibility_message)
if __name__ == "__main__": main()

#3.
def main():
    trainees = ["Nami", "Luffy", "Chopper", "Zoro"]
    scores = [99, 80, 76, 100]

    print("Trainees:")
    for trainee in trainees:
        print(trainee)

    print("\nScores:")
    for score in scores:
        print(score)
if __name__ == "__main__": main()

#4.
import pandas as pd
def main():
    data = {
        "Name": ["Nami", "Luffy", "Chopper", "Zoro"],
        "Score": [99, 80, 76, 100]
    }
    df = pd.DataFrame(data)
    df.set_index("Name", inplace=True)
    print(df)
if __name__ == "__main__": main()

#5.
import pandas as pd
df = pd.read_csv('/content/Project-Andrew_Phipps.csv')
df.head()
print(df.info())
print(df.describe())

import matplotlib.pyplot as plt
plt.plot(df['Year'], df['Student Population'], marker='o')
plt.title('Student Population Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Student Population')
plt.xticks(df['Year'], rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
plt.scatter(df['Retention Rate'], df['Graduation Rate'], s=80)
plt.title('Retention Rate vs Graduation Rate', fontsize=16, fontweight='bold')
plt.xlabel('Retention Rate')
plt.ylabel('Graduation Rate')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

#6.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('exercise')
print(df.describe())
print(df.info())

plt.figure(figsize=(10, 5))
df['pulse'].plot(kind='hist', bins=10)
plt.title("Pulse Distribution")
plt.xlabel("Pulse")
plt.ylabel("Frequency")
plt.show()
```
