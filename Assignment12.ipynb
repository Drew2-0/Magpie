```python
import seaborn as sns
import pandas as pd
from pandas.core.computation.check import NUMEXPR_INSTALLED
from matplotlib import pyplot as plt

# Load built-in dataset
df = sns.load_dataset('titanic')  # Example dataset
# Step 3
print(df.describe())
print(df.info())
# Step 4

sns.countplot(x='survived', data=df)
plt.title('Survival Count on Titanic', fontweight='bold', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')

plt.show()

plt.hist(df['age'].dropna())
plt.title('Age Distribution on the Titanic', fontweight='bold', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()

plt.scatter(df['age'], df['fare'])
plt.title('Age vs Fare on the Titanic', fontweight='bold', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
#Step 5
print("Over half of the passangers on the titanic died with the count beign rought 550 who died and 350 who survivied. The most common age group on the Titanic where 20 year olds who were clostly followed by 30 year olds. The most common fare price was between 0-25 dollors with the most expensive going for 500. ")
```
